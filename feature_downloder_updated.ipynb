### This is a program to download the gene features with Patric ID programmatically

```python
# Read the data
import pandas as pd

df = pd.read_excel("a4ed1601ccb774eaf3b9c5f44b1a6bc3_JCM.01260-18-s0002.xlsx", skiprows=1)
df
```

```python
# Find out the list of non repeating patrid IDs
pid_list = list(set(df["PATRIC ID"]))
len(pid_list)
```

#### Start download

```python
import os
import os.path

# Make a directory named "Features in the root directory"
os.system("mkdir Features")
count = 0

# Loop through the list of unique Patrid IDs and download the sequence with p3 api
for pid in pid_list:
    file_name = "_".join(str(pid).split("."))
    filename_full = "Features/features_{}.fasta".format(file_name)
    
    os.system("p3-genome-fasta --feature {} >{}".format(pid, filename_full))
    
    #Check if the downloaded file is empty
    if os.path.getsize(filename_full) > 0:
        count += 1
    else:
        os.system("rm {}".format(filename_full))
    
    # Count can be varied
    if count > 500:
        break
    print(f"{file_name} downloaded, count: {count}")
```
